```python
# system variables
alpha1 = sp.Symbol('alpha_1')  # servo1 angle
alpha2 = sp.Symbol('alpha_2')  # servo2 angle
alpha3 = sp.Symbol('alpha_3')  # servo3 angle
alpha = sp.Matrix([
    [alpha1],
    [alpha2],
    [alpha3]
])

# Coordinates
Ppx = sp.Symbol('P^p_x')  # x coord of ball in the platform frame 
Ppy = sp.Symbol('P^p_y')  # y coord of ball in the platform frame  
Ppz = sp.Symbol('P^p_z')  # z coord of ball in the platform frame  
Pp = sp.Matrix([
    [Ppx],
    [Ppy],
    [Ppz]
])
```

```python
# formulas

def rotation(phi, theta):  # no yaw movement
  # determines the rotational matrix
  Rx = sp.Matrix([  # roll: phi
      [1, 0, 0], 
      [0, sp.cos(phi), -sp.sin(phi)],
      [0, sp.sin(phi), sp.cos(phi)]
  ])
  Ry = sp.Matrix([  # pitch: theta
      [sp.cos(theta), 0, sp.sin(theta)], 
      [0, 1, 0],
      [-sp.sin(theta), 0, sp.cos(theta)]
  ])
  R = Rx@Ry
  return R

def transformation(R, T, p):
  # transforms point p by (R, T)
  if R.shape[0] == T.shape[0]:
    if R.shape[1] != T.shape[1]:
      T1 = T
      for i in range(R.shape[1]-T.shape[1]):
        T = sp.Matrix.hstack(T, T1)
    return R@p + T  # ASSUMPTION: THE ORIGIN OF THE PLATFORM FRAME DOES NOT MOVE
```

```python
# TODO
def pid(error, error_dot, error_sum):
  val = 0.012
  Kp = [0.001, 0.001] #kp cant be bigger than [ 0.00985, 0.0136] 
  Kd = [0.0001, 0.0001] #kp 
  Ki = [0.000001, 0.000001] #ki
  target_phi = -Kp[0]*error[1]-Kd[0]*error_dot[1]-Ki[0]*error_sum[1]  # y distance error requires x axis rotation - "-" to counteract ball
  target_theta = Kp[1]*error[0]+Kd[1]*error_dot[0]+Ki[1]*error_sum[0]  #x distance error requires y axis rotation
  
  #Set saturation angles for theta (y axis rotation) and phi (x axis rotation)
  target_theta = min(target_theta, np.arctan2(2*a, R*(np.cos(np.pi/3)+1))) #min of upper bound
  target_theta = max(target_theta, np.arctan2(-2*a, R*(np.cos(np.pi/3)+1))) # max of lower bound

  target_phi = min(target_phi, np.arctan2(2*a,2*R*np.sin(np.pi/3))) 
  target_phi = max(target_phi, np.arctan2(-2*a,2*R*np.sin(np.pi/3))) 

  return target_phi, target_theta
```

```python
# TODO
def map_image_to_platform(img_x, img_y):
  platform_x = img_x  # replace later
  platform_y = img_y  # replace later
  Pp = sp.Matrix([  # ASSUMPTION: THE TILTED ANGLE DOESN'T AFFECT THE POINT OF VIEW
      [platform_x],
      [platform_y],
      [0]
  ])
  return Pp

def find_next_q(Pp):
  # get target q based on ball position (Pp) in platform frame, origin (Op) of the platform, and the q in platform (Prod)
  error = Op - Pp  # 3x3 matrix
  target_phi, target_theta = pid(error)  # pid target alpha vs feedfoward target alpha
  rot_mat = rotation(target_phi, target_theta)
  T = sp.Matrix([[0], [0], [h]])
  # find the coordinates of the rod ends in base frame
  # print(T, Prod)
  target_q = transformation(rot_mat, T, Prod)  
  return target_q

def find_next_alpha(target_q):
  alpha = sp.Matrix([
    [alpha1],
    [alpha2],
    [alpha3]
  ])
  # step 2-1
  B = sp.Matrix([
      [R*sp.cos(beta1), R*sp.cos(beta2), R*sp.cos(beta3)],
      [R*sp.sin(beta1), R*sp.sin(beta2), R*sp.sin(beta3)],
      [0, 0, 0]
  ]) #+ Ob  # add the origin of the base frame (should just be [0,0,0])
  l = target_q-B
  # print(target_q)
  # step 2-2
  for i in range(3):
    l_norm = l[:,i].norm()
    # print(l_norm**2, (s**2 - a**2))
    Li = l_norm**2 - (s**2 - a**2)
    xi = l[0,i]
    yi = l[1,i]
    zi = l[2,i]
    Mi = 2*a*zi
    Ni = 2*a*sp.cos(betas[i])*xi+sp.sin(betas[i])*yi
    # print(Li, Mi, Ni)
    alpha[i] = sp.asin(Li/sp.sqrt(Mi**2+Ni**2)-sp.atan2(Ni, Mi))
  alpha = sp.simplify(alpha)
  # SUBSTITUTE CONSTANTS HERE by alpha.subs([(variable, value),...])
  return alpha
```

```python
def simulate(ball_x, ball_y):
  Pp = map_image_to_platform(ball_x, ball_y)
  target_q = find_next_q(Pp)
  target_alpha = find_next_alpha(target_q)
  return target_alpha
```

```python
# Simulate with a simulated ball on platform
g = 9810
Ts = 0.07 #sampling time average from OpenCV Serial limitation
xprev = 50
yprev = 50
xsim = []
ysim = []
esim = []
t = np.arange(0, 10, Ts)
N = 5  # horizon
for idx, i in enumerate(t):
  Pp = np.array([  # ASSUMPTION: THE TILTED ANGLE DOESN'T AFFECT THE POINT OF VIEW
  [xprev],
  [yprev],
  [0]])
  # find errors
  error = -Pp
  esim.append(error)
  e_dot = (esim[idx] - esim[idx-1])/Ts if idx != 0 else esim[idx]/Ts # discrete derivative of the error
  e_sum = 1/N*np.sum([esim[n] if n < len(esim) and n >= 0 else 0 for n in np.arange(idx-N+1, idx+1)], axis=0)  # moving sum of an interval
  #Call pid(error)
  target_phi, target_theta = pid(error, e_dot, e_sum)
  target_phi = float(target_phi)
  target_theta = float(target_theta)
  #Calculate new X
  xnew = Ts*i*g*np.sin(target_theta)+ xprev
  #Calculate new Y
  ynew = Ts*i*g*np.sin(-target_phi)+ yprev
  # print(i, target_phi, target_theta)
  #Append xynew to sim array
  xsim.append(xnew)
  ysim.append(ynew)
  #Update xyprev
  xprev = xnew
  yprev = ynew

fig, ax = plt.subplots(1,2, figsize=(10, 6))
ax[0].plot(t, xsim, '.-')
ax[0].set_xlim(0, 2)
ax[0].set_ylim(-70, 70)
ax[1].plot(t, ysim, '.-')
ax[1].set_xlim(0, 2)
ax[1].set_ylim(-70, 70)
```
